```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
```

```python
inp = data.copy()
x = inp[[heads[11],heads[6],heads[4],heads[8]]]
#x = inp[[heads[11]]]

y = inp[heads[13]]

print(x)

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)

a=data[heads[13]].value_counts()
print(a.iloc[0]/5000)
```

```python
def ana(A,B,C,D):
    print("accuracy           ",(A+D)/(A+B+C+D))
    if A == 0:
        A = 1
        print("error")
    if B == 0:
        B = 1
        print("error")
    if C == 0:
        C = 1
        print("error")
    if D == 0:
        D = 1
        print("error")
    
    pre1 = A/(A+C)
    pre2 = D/(D+B)
    recall1 = A/(A+B)
    recall2 = C/(C+D)
    print("precision for 0:   ",pre1)
    print("precision for 1:   ",pre2)
    print("")
    print("Recall for 0:      ",recall1)
    print("Recall for 1:      ",recall2)
    print("")
    f11 = (2*pre1*recall1)/(pre1 + recall1)
    f12 = (2*pre2*recall2)/(pre2 + recall2)

    print("f1 score for 0:    ",f11)
    print("f1 score for 1:    ",f12)
```

```python
 from sklearn.metrics import confusion_matrix as cnfmat
```

```python
gnb = GaussianNB()
y_pred1 = gnb.fit(x_train, y_train).predict(x_test)

print("mean of predicted values :",y_pred1.mean())
print("accuracy:",gnb.score(x_test,y_test))
m = cnfmat(y_test,y_pred1)
print(m)
print()
A = m[0,0]
B = m[1,0]
C = m[0,1]
D = m[1,1]
print(A,B,C,D)
ana(A,B,C,D)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
ada = GradientBoostingClassifier(n_estimators=100, learning_rate=.1,max_depth=1, random_state=42)
y_pred2 = ada.fit(x_train, y_train).predict(x_test)

print("mean of predicted values :",y_pred2.mean())
print("accuracy:",ada.score(x_test,y_test))
m = cnfmat(y_test,y_pred2)
print()
print(m)
A = m[0,0]
B = m[1,0]
C = m[0,1]
D = m[1,1]
print(A,B,C,D)
ana(A,B,C,D)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3)
#knn = KNeighborsClassifier(n_neighbors=3)
model = knn.fit(x_train, y_train)
y_pred3 = model.predict(x_test)

print(y_pred3.mean())
print(knn.score(x_test,y_test))

m = cnfmat(y_test,y_pred3)
print(m)
A = m[0,0]
B = m[1,0]
C = m[0,1]
D = m[1,1]
print(A,B,C,D)
ana(A,B,C,D)
```

```python
inp = data.copy()
inp["knn"] = model.predict(x)

inp["CCAvg"] = (inp["CCAvg"] - inp["CCAvg"].mean() ) / inp["CCAvg"].std()
inp["Mortgage"] = (inp["Mortgage"] - inp["Mortgage"].mean() ) / inp["Mortgage"].std()
inp["ZIP Code"] = (inp["ZIP Code"] - inp["ZIP Code"].mean() ) / inp["ZIP Code"].std()
inp["o1"] = pd.get_dummies(inp[heads[11]])[0]
inp["o2"] = pd.get_dummies(inp[heads[11]])[1]
inp
```

```python
x = inp[["ZIP Code","CCAvg","Mortgage","o1","o2","knn"]]

y = pd.get_dummies(inp[heads[13]])


x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.1, random_state=42)
print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)
print(x_train)
#a=data[heads[13]].value_counts()
#print(a.iloc[0]/5000)
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
```

```python
class Net(nn.Module):

    def __init__(self):
        super(Net, self).__init__()
        self.fc1 = nn.Linear(6, 6)
        self.fc2 = nn.Linear(6, 6)
        self.fc3 = nn.Linear(6, 2)

    def forward(self, x):
        x = torch.tanh(self.fc1(x))
        x = F.elu(self.fc2(x))
        x = F.elu(self.fc3(x))
        return x
        
```

```python
net = Net()
print(net)
testing = torch.tensor([list(x_train.iloc[0])])
l = torch.tensor([list(y_train.iloc[0])])
print(l)
print(testing)
net(testing)
```

```python
import torch.optim as optim

criterion = nn.CrossEntropyLoss(weight = torch.tensor([1.0,0.4]),reduction="sum")
optimizer = optim.Adam(net.parameters())
```

```python
import time
st = time.time()
dots = []
for epoch in range(5):  # loop over the dataset multiple times
    rn = 0
    for i in range(len(x_train)):
        xi = torch.tensor([list(x_train.iloc[i])])
        yi = torch.tensor([list(y_train.iloc[i])[0]])
        optimizer.zero_grad()

        outputs = net(xi)
        #outputs=outputs.reshape(1,2)
        loss = criterion(outputs, yi)
        loss.backward()
        optimizer.step()

        rn += loss.item()
        
        if i % 1500 == 1499 :    # print every 2000 mini-batches
            print('[%d, %5d] loss: %f' %
                  (epoch + 1, i + 1, rn/1499))
            
    dots.append(rn/len(x_test))
print('Finished Training')
plt.plot(dots)
plt.show()
print(time.time() - st)
```

```python

A = 0
B = 0
C = 0
D = 0
idk = 0

#print(test.head(50))
print(y_test[y_test.columns[0]].value_counts())
for i in range(len(x_test)):
    xi = torch.tensor([list(x_test.iloc[i])])
    yi = torch.tensor([list(y_test.iloc[i])[0]])

    out = net(xi)
    #print(out[:,0] > out[:,1])
    if yi == 0 and out[:,0] > out[:,1]: # pre was 0 and ans was 0 
        A += 1
    elif yi == 1 and out[:,0] > out[:,1]: # pre was 1 but ans was 0
        B += 1
    elif yi == 0 and out[:,0] < out[:,1]: # pre was 0 but ans was 1
        C += 1
    elif yi == 1 and out[:,0] < out[:,1]: # pre was 1 and ans was 1
        D += 1
    else:
        idk += 1

print(idk)

stats = pd.DataFrame([[A, B], [C, D]],index=("0","1"),columns=["0","1"])
print(stats)
ana(A,B,C,D)
```
